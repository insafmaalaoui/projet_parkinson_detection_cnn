```python
# Cell 1
import matplotlib.pyplot as plt
import numpy as np
from src.data.preprocessing import load_full_dataset

X, y, bad = load_full_dataset()
print("X shape:", X.shape)
print("y shape:", y.shape)
print("Bad files example:", bad[:5])
```

```python
# Cell 2
(unique, counts) = np.unique(y, return_counts=True)
print("Distribution:", dict(zip(unique, counts)))

# afficher quelques images
n_show = min(10, X.shape[0])
fig, axes = plt.subplots(2, 5, figsize=(12,5))
idxs = np.random.choice(X.shape[0], n_show, replace=False)
for ax, idx in zip(axes.flatten(), idxs):
    ax.imshow(X[idx])
    ax.set_title(f"Label: {y[idx]}")
    ax.axis("off")
plt.tight_layout()
```
